```python
def remove_spaces_between_words(genres):
    return genres.replace(' ', '')

# Apply the function to 'genres' column
movies['genres'] = movies['genres'].apply(remove_spaces_between_words)
movies['keywords'] = movies['keywords'].apply(remove_spaces_between_words)
```

```python
movies.head(2)
```

```python
d=movies[ 'original_language'].unique()
print(d)
```

```python
movies['adult']=movies['adult'].astype(str)
```

```python
movies[ 'overview']=movies[ 'overview'].apply(lambda x:x.split())
movies[ 'genres']=movies[ 'genres'].apply(lambda x:x.split())
movies['keywords']=movies['keywords'].apply(lambda x:x.split())
movies['adult']=movies['adult'].apply(lambda x:x.split())
movies[ 'original_language']=movies[ 'original_language'].apply(lambda x:x.split())
movies.head(2)
```

```python
movies['tags']= movies['adult'] + movies[ 'original_language'] + movies[ 'overview'] + movies[ 'genres'] + movies['keywords'] 
```

```python
movies.head(2)
```

```python
new_movie=movies[['id', 'title', 'tags']]
new_movie.head(2)
```

```python
new_movie['tags']=new_movie['tags'].apply(lambda x:" ".join(x))
```

```python
new_movie['tags']=new_movie['tags'].apply(lambda x:x.lower())
```

```python
new_movie.head(2)
```

```python
import nltk
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
    y=[]
    for i in text.split():
        y.append(ps.stem(i))
        
    return " ".join(y)
```

```python
new_movie['tags'] = new_movie['tags'].apply(stem)
new_movie.head(2)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=3000, stop_words='english')
```

```python
new_movie.columns
```

```python
vector=cv.fit_transform(new_movie['tags']).toarray()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity=cosine_similarity(vector)
```

```python
sorted(list(enumerate(similarity[0])),reverse=True,key=lambda x:x[1])[1:6]
```

```python
def recommend(movie):
    movie_index= new_movie[new_movie['title']==movie].index[0]
    dist=similarity[movie_index]
    movie_list=sorted(list(enumerate(dist)),reverse=True,key=lambda x:x[1])[1:6]
    
    for i in movie_list:
        print(new_movie.iloc[i[0]].title)
```

```python
recommend("Inception")
```
